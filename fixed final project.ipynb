```python
#fill the hat with random numbers of tickets of each color
import random
balance=200
t1=[]
thing=True
while thing:
    if len(t1)<100:
        i=random.randrange(1,6)
        if i==1:
            t1.append('red')
        elif i==2:
            t1.append('blue')
        elif i==3:
            t1.append('green')
        elif i==4:
            t1.append('yellow')
        elif i==5:
            t1.append('orange')
        elif i==6:
            t1.append('purple')
    else:
        break
a1=t1.count('red')
a2=t1.count('blue')
a3=t1.count('green')
a4=t1.count('yellow')
a5=t1.count('orange')
a6=t1.count('purple')
#define lottery class to allow for drawing and seeing probability
class lottery():
    def __init__(self,red,blue,green,yellow,orange,purple):
        self.red=t1.count('red')
        self.blue=t1.count('blue')
        self.green=t1.count('green')
        self.yellow=t1.count('yellow')
        self.orange=t1.count('orange')
        self.purple=t1.count('purple')
#allow player to see the chances of drawing the chosen color
    def redchance(self):
        print(f'there are {self.red} red tickets in the hat')
        print(f'the chances of drawing a red ticket are {self.red/(len(t1))*100}%')
        print(f'the prize for correctly guessing red is ${100-self.red/(len(t1))*100}')
    def bluechance(self):
        print(f'there are {self.blue} blue tickets in the hat')
        print(f'the chances of drawing a blue ticket are {self.blue/(len(t1))*100}%')
        print(f'the prize for correctly guessing blue is ${100-self.blue/(len(t1))*100}')
    def greenchance(self):
        print(f'there are {self.green} green tickets in the hat')
        print(f'the chances of drawing a green ticket are {self.green/(len(t1))*100}%')
        print(f'the prize for correctly guessing green is ${100-self.green/(len(t1))*100}')
    def yellowchance(self):
        print(f'there are {self.yellow} yellow tickets in the hat')
        print(f'the chances of drawing a yellow ticket are {self.yellow/(len(t1))*100}%')
        print(f'the prize for correctly guessing yellow is ${100-self.yellow/(len(t1))*100}')
    def orangechance(self):
        print(f'there are {self.orange} orange tickets in the hat')
        print(f'the chances of drawing an orange ticket are {self.orange/(len(t1))*100}%')
        print(f'the prize for correctly guessing orange is ${100-self.orange/(len(t1))*100}')
    def purplechance(self):
        print(f'there are {self.purple} purple tickets in the hat')
        print(f'the chances of drawing a purple ticket are {self.purple/(len(t1))*100}%')
        print(f'the prize for correctly guessing purple is ${100-self.purple/(len(t1))*100}')
#allow player to draw a ticket
    def reddraw(self):
        print('you drew a red ticket')
        t1.remove('red')
    def bluedraw(self):
        print('you drew a blue ticket')
        t1.remove('blue')
    def greendraw(self):
        print('you drew a green ticket')
        t1.remove('green')
    def yellowdraw(self):
        print('you drew a yellow ticket')
        t1.remove('yellow')
    def orangedraw(self):
        print('you drew an orange ticket')
        t1.remove('orange')
    def purpledraw(self):
        print('you drew a purple ticket')
        t1.remove('purple')
#set up the game
play=True
print('welcome to the raffel\n100 tickets of different colors have been placed into a hat\nyou can either draw a ticket out of the hat for $20 or see how many tickets of a particular color are still in the hat and the chances of you drawing it\nguessing correctly gives you a prize')
while play:
    game=input(f"you currently have ${balance} remaining. would you like to play? yes/no: ")
    if game=='yes':
        game1=lottery(a1,a2,a3,a4,a5,a6)
        try:
            option=int(input('to draw a ticket, press 1. to see the chances, press 2.'))
        except:
            print('enter either 1 or 2')
        else:
            if option==1:
                if balance<20:
                    print('you cannot afford to draw a ticket')
                else:
                    #let players pick a color
                    balance-=20
                    c=input('pick a color of ticket you hope to draw:\noptions are red, blue, green, yellow, orange, or purple ')
                    if c!='red' and c!='blue' and c!='green' and c!='yellow' and c!='orange' and c!='purple':
                        print('select a color')
                        balance+=20
                    else:
                        d=random.choice(t1)
                        if d=='red':
                            game1.reddraw()
                            e='red'
                        elif d=='blue':
                            game1.bluedraw()
                            e='blue'
                        elif d=='green':
                            game1.greendraw()
                            e='green'
                        elif d=='yellow':
                            game1.yellowdraw()
                            e='yellow'
                        elif d=='orange':
                            game1.orangedraw()
                            e='orange'
                        elif d=='purple':
                            game1.purpledraw()
                            e='purple'
                        else:
                            print('error')
                        if c==e:
                            print('you have guessed correctly')
                            if c=='red':
                                balance+=100-(a1/(len(t1))*100)
                                print(f'you have won ${100-(a1/(len(t1))*100)}')
                            elif c=='blue':
                                balance+=100-(a1/(len(t1))*100)
                                print(f'you have won ${100-(a2/(len(t1))*100)}')
                            elif c=='green':
                                balance+=100-(a1/(len(t1))*100)
                                print(f'you have won ${100-(a3/(len(t1))*100)}')
                            elif c=='yellow':
                                balance+=100-(a1/(len(t1))*100)
                                print(f'you have won ${100-(a4/(len(t1))*100)}')
                            elif c=='orange':
                                balance+=100-(a1/(len(t1))*100)
                                print(f'you have won ${100-(a5/(len(t1))*100)}')
                            elif c=='purple':
                                balance+=100-(a1/(len(t1))*100)
                                print(f'you have won ${100-(a6/(len(t1))*100)}')
                        else:
                            print('you have guessed incorrectly')
            elif option==2:
                if balance<1:
                    print('you cannot afford to see the chances')
                else:
                    #allow players to see the chances
                    balance-=1
                    chance=input('select a color you would like to see the chances for:\noptions are red, blue, green, yellow, orange, or purple  ')
                    if chance=='red':
                        game1.redchance()
                    elif chance=='blue':
                        game1.bluechance()
                    elif chance=='green':
                        game1.greenchance()
                    elif chance=='yellow':
                        game1.yellowchance()
                    elif chance=='orange':
                        game1.orangechance()
                    elif chance=='purple':
                        game1.purplechance()
                    #continue the program if they give invalid outputs
                    else:
                        print('select one of the colors')
                        balance+=1
            else:
                print('select another option')
#end the game if a player inputs 'no'
    elif game=='no':
        break
#make the user reenter answer if they don't say yes or no
    else:
        print('select either yes or no')
```
